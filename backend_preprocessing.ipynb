```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('dataset/cocktails.csv')
df2 = pd.read_csv('dataset/cocktails_db.csv')
df = pd.concat([df1, df2], axis=0)

df['Cocktail Name'] = df['Cocktail Name'].str.upper()
df.drop_duplicates(subset='Cocktail Name', inplace=True)
df.fillna('', inplace=True)
df['All Ingredients'] = df['Ingredients'] + ',' + df['Garnish']
df.drop(columns=['Bartender', 'Location', 'Bar/Company', 'Glassware', 'Notes'], inplace=True)
df.reset_index()
```
